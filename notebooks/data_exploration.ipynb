```python
import numpy as np
import pandas as pd
```

```python
data_path = '../downloaded_data_2019_02_01/planes.csv'
df = pd.read_csv(data_path, parse_dates=['first_flight'])

# Parse dates to datetime
df['first_flight'] = pd.to_datetime(df['first_flight'], errors='coerce', format="%d/%m/%Y") 
df.set_index(np.arange(df.shape[0]))
#df.index.name = 'id'
```
